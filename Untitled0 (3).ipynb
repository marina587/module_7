```python
import numpy as np
import matplotlib.pyplot as plt
import torch
import torch.nn as nn
import torch.optim as optim
from torch.utils.data import Dataset, DataLoader
from tqdm import tqdm
import torch.nn.functional as F

```

```python
USE_RGB = True  # Установите True для RGB, False для оттенков серого
print(f"🔹 Активный режим: {'RGB (3 канала)' if USE_RGB else 'Grayscale (1 канал)'}")

device = torch.device('cuda' if torch.cuda.is_available() else 'cpu')
print(f"🔹 Используем устройство: {device}")
```

```python
# 1. ГЕНЕРАЦИЯ ДАННЫХ (из вашего ноутбука)
# ==========================================

class BaseDataProvider(object):
    channels = 1
    n_class = 2

    def __init__(self, a_min=None, a_max=None):
        self.a_min = a_min if a_min is not None else -np.inf
        self.a_max = a_max if a_max is not None else np.inf

    def _load_data_and_label(self):
        data, label = self._next_data()
        train_data = self._process_data(data)
        labels = self._process_labels(label)
        train_data, labels = self._post_process(train_data, labels)
        nx = train_data.shape[1]
        ny = train_data.shape[0]
        return train_data.reshape(1, ny, nx, self.channels), labels.reshape(1, ny, nx, self.n_class)

    def _process_labels(self, label):
        if self.n_class == 2:
            nx = label.shape[1]
            ny = label.shape[0]
            labels = np.zeros((ny, nx, self.n_class), dtype=np.float32)
            if label.dtype != 'bool':
                label = label.astype(bool)
            labels[..., 1] = label
            labels[..., 0] = ~label
            return labels
        return label

    def _process_data(self, data):
        data = np.clip(np.fabs(data), self.a_min, self.a_max)
        data -= np.amin(data)
        if np.amax(data) != 0:
            data /= np.amax(data)
        return data

    def _post_process(self, data, labels):
        return data, labels

    def __call__(self, n):
        train_data, labels = self._load_data_and_label()
        nx = train_data.shape[1]
        ny = train_data.shape[2]
        X = np.zeros((n, nx, ny, self.channels))
        Y = np.zeros((n, nx, ny, self.n_class))
        X[0] = train_data
        Y[0] = labels
        for i in range(1, n):
            train_data, labels = self._load_data_and_label()
            X[i] = train_data
            Y[i] = labels
        return X, Y

class GrayScaleDataProvider(BaseDataProvider):
    channels = 1
    n_class = 2
    def __init__(self, nx, ny, **kwargs):
        super(GrayScaleDataProvider, self).__init__()
        self.nx = nx
        self.ny = ny
        self.kwargs = kwargs

    def _next_data(self):
        return create_image_and_label(self.nx, self.ny, **self.kwargs)

class RgbDataProvider(BaseDataProvider):
    channels = 3
    n_class = 2
    def __init__(self, nx, ny, **kwargs):
        super(RgbDataProvider, self).__init__()
        self.nx = nx
        self.ny = ny
        self.kwargs = kwargs

    def _next_data(self):
        data, label = create_image_and_label(self.nx, self.ny, **self.kwargs)
        return to_rgb(data), label

def create_image_and_label(nx, ny, cnt=10, r_min=5, r_max=50, border=92, sigma=20, rectangles=False):
    image = np.ones((nx, ny, 1))
    label = np.zeros((nx, ny, 3), dtype=bool)
    mask = np.zeros((nx, ny), dtype=bool)
    for _ in range(cnt):
        a = np.random.randint(border, nx - border)
        b = np.random.randint(border, ny - border)
        r = np.random.randint(r_min, r_max)
        h = np.random.randint(1, 255)
        y, x = np.ogrid[-a:nx - a, -b:ny - b]
        m = x * x + y * y <= r * r
        mask = np.logical_or(mask, m)
        image[m] = h
    label[mask, 1] = 1
    if rectangles:
        mask = np.zeros((nx, ny), dtype=bool)
        for _ in range(cnt // 2):
            a = np.random.randint(nx)
            b = np.random.randint(ny)
            r = np.random.randint(r_min, r_max)
            h = np.random.randint(1, 255)
            m = np.zeros((nx, ny), dtype=bool)
            m[a:a + r, b:b + r] = True
            mask = np.logical_or(mask, m)
            image[m] = h
        label[mask, 2] = 1
        label[..., 0] = ~(np.logical_or(label[..., 1], label[..., 2]))
    image += np.random.normal(scale=sigma, size=image.shape)
    image -= np.amin(image)
    image /= np.amax(image)
    if rectangles:
        return image, label
    else:
        return image, label[..., 1]

def to_rgb(img):
    img = img.reshape(img.shape[0], img.shape[1])
    img[np.isnan(img)] = 0
    img -= np.amin(img)
    img /= np.amax(img)
    blue = np.clip(4 * (0.75 - img), 0, 1)
    red = np.clip(4 * (img - 0.25), 0, 1)
    green = np.clip(44 * np.fabs(img - 0.5) - 1., 0, 1)
    rgb = np.stack((red, green, blue), axis=2)
    return rgb
```

```python
# 2. ПОДГОТОВКА ДАННЫХ И РАЗБИЕНИЕ
# ==========================================

class CircleDataset(Dataset):
    def __init__(self, images, masks):
        self.images = images
        self.masks = masks

    def __len__(self):
        return len(self.images)

    def __getitem__(self, idx):
        img = self.images[idx]
        mask = self.masks[idx]

        # Преобразование HWC -> CHW для изображения
        img_tensor = torch.from_numpy(img).float().permute(2, 0, 1)

        # Маска: берем канал 1 (кружочки) и ОБЯЗАТЕЛЬНО добавляем размерность канала (1, H, W)
        mask_tensor = torch.from_numpy(mask[..., 1]).float().unsqueeze(0)

        return img_tensor, mask_tensor

def generate_splits(num_train=500, num_test=100, num_val=100, nx=256, ny=256):
    # Выбор генератора на основе флага
    ProviderClass = RgbDataProvider if USE_RGB else GrayScaleDataProvider
    generator = ProviderClass(nx, ny, cnt=20)

    print("Генерация Train набора...")
    x_train, y_train = generator(num_train)
    print("Генерация Test набора...")
    x_test, y_test = generator(num_test)
    print("Генерация Validation набора...")
    x_val, y_val = generator(num_val)

    return (x_train, y_train), (x_test, y_test), (x_val, y_val)

data_train, data_test, data_val = generate_splits()

train_dataset = CircleDataset(*data_train)
test_dataset = CircleDataset(*data_test)
val_dataset = CircleDataset(*data_val)

batch_size = 8
train_loader = DataLoader(train_dataset, batch_size=batch_size, shuffle=True, num_workers=2)
test_loader = DataLoader(test_dataset, batch_size=batch_size, shuffle=False, num_workers=2)
val_loader = DataLoader(val_dataset, batch_size=batch_size, shuffle=False, num_workers=2)
```

```python
# 3. АРХИТЕКТУРА UNET
# ==========================================

class DoubleConv(nn.Module):
    def __init__(self, in_channels, out_channels):
        super().__init__()
        self.conv = nn.Sequential(
            nn.Conv2d(in_channels, out_channels, 3, 1, 1, bias=False),
            nn.BatchNorm2d(out_channels),
            nn.ReLU(inplace=True),
            nn.Conv2d(out_channels, out_channels, 3, 1, 1, bias=False),
            nn.BatchNorm2d(out_channels),
            nn.ReLU(inplace=True),
        )
    def forward(self, x):
        return self.conv(x)

class UNET(nn.Module):
    def __init__(self, in_channels=1, out_channels=1, features=[64, 128, 256, 512]):
        super().__init__()
        self.ups = nn.ModuleList()
        self.downs = nn.ModuleList()
        self.pool = nn.MaxPool2d(kernel_size=2, stride=2)

        for feature in features:
            self.downs.append(DoubleConv(in_channels, feature))
            in_channels = feature

        self.bottleneck = DoubleConv(features[-1], features[-1] * 2)

        for feature in reversed(features):
            self.ups.append(nn.ConvTranspose2d(feature * 2, feature, kernel_size=2, stride=2))
            self.ups.append(DoubleConv(feature * 2, feature))

        self.final_conv = nn.Conv2d(features[0], out_channels, kernel_size=1)

    def forward(self, x):
        skip_connections = []
        for down in self.downs:
            x = down(x)
            skip_connections.append(x)
            x = self.pool(x)
        x = self.bottleneck(x)
        skip_connections = skip_connections[::-1]

        for idx in range(0, len(self.ups), 2):
            x = self.ups[idx](x)
            skip_connection = skip_connections[idx // 2]
            if x.shape != skip_connection.shape:
                x = F.resize(x, size=skip_connection.shape[2:])
            concat_skip = torch.cat((skip_connection, x), dim=1)
            x = self.ups[idx + 1](concat_skip)

        return self.final_conv(x)

# Инициализация модели с учётом режима
in_channels = 3 if USE_RGB else 1
model = UNET(in_channels=in_channels, out_channels=1).to(device)
```

```python
# 4. ОБУЧЕНИЕ
def iou_score(outputs, labels, threshold=0.5):
    outputs = (outputs > threshold).float()
    intersection = (outputs * labels).sum()
    union = outputs.sum() + labels.sum() - intersection
    return (intersection + 1e-6) / (union + 1e-6)

criterion = nn.BCEWithLogitsLoss()
optimizer = optim.Adam(model.parameters(), lr=1e-4)

num_epochs = 20
train_losses, test_losses, test_ious = [], [], []

print("\n🚀 Начало обучения...")
for epoch in range(num_epochs):
    # --- TRAIN ---
    model.train()
    running_train_loss = 0.0
    pbar = tqdm(train_loader, desc=f"Epoch {epoch+1}/{num_epochs} [Train]")
    for inputs, masks in pbar:
        inputs, masks = inputs.to(device), masks.to(device)
        optimizer.zero_grad()
        outputs = model(inputs)
        loss = criterion(outputs, masks)
        loss.backward()
        optimizer.step()
        running_train_loss += loss.item()
        pbar.set_postfix({'Loss': loss.item()})
    train_losses.append(running_train_loss / len(train_loader))

    # --- TEST ---
    model.eval()
    running_test_loss = 0.0
    current_test_iou = 0.0
    num_test_batches = 0
    with torch.no_grad():
        pbar = tqdm(test_loader, desc=f"Epoch {epoch+1}/{num_epochs} [Test] ", leave=False)
        for inputs, masks in pbar:
            inputs, masks = inputs.to(device), masks.to(device)
            outputs = model(inputs)
            loss = criterion(outputs, masks)
            running_test_loss += loss.item()
            iou = iou_score(outputs, masks)
            current_test_iou += iou.item()
            num_test_batches += 1
    test_losses.append(running_test_loss / len(test_loader))
    test_ious.append(current_test_iou / num_test_batches if num_test_batches > 0 else 0.0)

    print(f"Epoch {epoch+1} | Train Loss: {train_losses[-1]:.4f} | Test Loss: {test_losses[-1]:.4f} | Test IoU: {test_ious[-1]:.4f}")

# ===================
```

```python
# 5. ГРАФИК LOSS
# ==========================================

plt.figure(figsize=(10, 5))
plt.plot(train_losses, label='Train Loss', marker='o')
plt.plot(test_losses, label='Test Loss', marker='o')
plt.title('Loss during training')
plt.xlabel('Epoch')
plt.ylabel('Loss')
plt.legend()
plt.grid(True)
plt.show()
```

```python
# 6. ОЦЕНКА НА ВАЛИДАЦИИ & ВИЗУАЛИЗАЦИЯ
print("\
📊 Оценка качества на Validation выборке...")
model.eval()
total_val_iou = 0.0
num_val_batches = 0

with torch.no_grad():
    for inputs, masks in tqdm(val_loader, desc="Evaluating"):
        inputs, masks = inputs.to(device), masks.to(device)
        outputs = model(inputs)
        probs = torch.sigmoid(outputs)
        iou = iou_score(probs, masks)
        total_val_iou += iou.item()
        num_val_batches += 1

mean_val_iou = total_val_iou / num_val_batches
print(f"✅ Средний IoU на валидации: {mean_val_iou:.4f}")

# Визуализация результатов
print("\
🖼️ Визуализация результатов (Input | Ground Truth | Prediction)")
fig, axes = plt.subplots(3, 3, figsize=(15, 15))
model.eval()

with torch.no_grad():
    for i in range(3):
        inputs, masks = val_dataset[i]
        inputs_batch = inputs.unsqueeze(0).to(device)
        outputs = model(inputs_batch)

        # ✅ FIX: Use .squeeze() instead of .squeeze(0) to remove both batch & channel dims
        # Resulting shape: (256, 256) -> perfectly valid for grayscale imshow
        probs = torch.sigmoid(outputs).cpu().squeeze()

        # Input
        if in_channels == 1:
            img_display = inputs.squeeze(0)
            axes[i, 0].imshow(img_display, cmap='gray')
        else:
            img_display = inputs.permute(1, 2, 0)
            axes[i, 0].imshow(img_display)
        axes[i, 0].set_title("Input Image")
        axes[i, 0].axis('off')

        # Ground Truth
        axes[i, 1].imshow(masks.squeeze(0), cmap='gray')
        axes[i, 1].set_title("Ground Truth")
        axes[i, 1].axis('off')

        # Prediction
        axes[i, 2].imshow(probs, cmap='gray')
        axes[i, 2].set_title("Prediction")
        axes[i, 2].axis('off')

plt.tight_layout()
plt.show()
```
